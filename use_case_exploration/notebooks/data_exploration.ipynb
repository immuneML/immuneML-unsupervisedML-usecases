```python
# plot the distribution of epitopes for each dataset
import matplotlib.pyplot as plt

for dataset in vdjdb_datasets:
    dataset_path = os.path.join(vdjdb_path, dataset)
    dataset_df = pd.read_csv(dataset_path, sep="\t")
    # filter for gene TRB
    dataset_df = dataset_df[dataset_df['Gene'] == 'TRB']
    print(f"Dataset: {dataset} ({len(dataset_df)} sequences)")
    epitope_counts = dataset_df['Epitope'].value_counts()
    plt.figure(figsize=(10, 6))
    epitope_counts.plot(kind='bar')
    plt.title(f"Distribution of Epitopes in {dataset}")
    plt.xlabel("Epitope")
    plt.ylabel("Count")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
